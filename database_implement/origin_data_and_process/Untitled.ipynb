```python
import pandas as pd
import sqlite3

def dataframe_to_db(dataframe, db_file, table_name):
    conn = sqlite3.connect(db_file)

    # 데이터프레임을 DB 테이블로 저장
    dataframe.to_sql(table_name, conn, if_exists='replace', index=True)

    conn.close()

# 사용 예시
db_file = 'food.db'
table_name = 'food'

dataframe_to_db(filtered_df, db_file, table_name)
```

```python
import sqlite3

def check_db(db_file, table_name):
    conn = sqlite3.connect(db_file)
    c = conn.cursor()

    # 테이블 정보 조회
    c.execute(f"SELECT sql FROM sqlite_master WHERE type='table' AND name='{table_name}'")
    create_table_query = c.fetchone()[0]

    # 테이블 컬럼 정보 추출
    start_index = create_table_query.index("(") + 1
    end_index = create_table_query.index(")")
    column_info = create_table_query[start_index:end_index]
    columns = [col.strip().split()[0] for col in column_info.split(",")]

    conn.close()

    return columns

# 사용 예시
db_file = 'food.db'
table_name = 'food'

columns = check_db(db_file, table_name)


print(columns)
```

```python
import sqlite3

def get_column_data_types(db_file, table_name):
    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()

    # 테이블 스키마 조회
    cursor.execute(f"PRAGMA table_info({table_name})")
    columns = cursor.fetchall()

    column_data_types = {}
    for column in columns:
        column_name = column[1]
        data_type = column[2]
        column_data_types[column_name] = data_type

    conn.close()

    return column_data_types

# 사용 예시
db_file = 'food.db'
table_name = 'food'

column_data_types = get_column_data_types(db_file, table_name)
for column, data_type in column_data_types.items():
    print(f"Column: {column}, Data Type: {data_type}")
```
